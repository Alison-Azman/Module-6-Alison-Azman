Object reusability

Object reusability means that a class can be written once and reused to create many objects.

Each object created from the class:

- is an instance of the same class

- shares the same structure and behavior

- can store different data values

This allows a single class definition to generate a variety of objects that all follow the same design.

A class can be created once and then used to make many objects.

All of these objects come from the same class, so they behave in the same way, even if their data is
different.




Privacy in Python

Python's approach to privacy is based on convention rather than strict enforcement.

In practice, it is possible to access most data inside an object.
However, Python programmers agree to follow naming conventions that indicate which data should be
treated as private.

Two common conventions are:

- Double underscore ( __ name)
Indicates that an attribute is intended to be more private.
Python uses name mangling to make accidental access harder.

- Single underscore (_name)
Indicates that an attribute is internal and should not be accessed directly outside the class.

These conventions communicate intent, helping programmers use objects correctly and safely.
