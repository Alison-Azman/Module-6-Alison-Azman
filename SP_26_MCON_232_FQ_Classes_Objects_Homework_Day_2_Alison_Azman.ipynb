## Big Reminders
A **class** is a blueprint for making objects.  
An **object** is one instance of that class.  
The constructor `__init__` gives each object its starting values.  
Instance attributes usually belong to one object.  
Class attributes belong to the class and are shared unless shadowed.  
Methods let objects do work.  
One object can receive another object as a parameter and use that object's data.  
If a method mutates an object, that mutation is still there when the method returns.  
If a method reassigns the parameter variable to a different object, the original caller's object does not get replaced.

## Task 9 — A `BankAccount` class

Create a class named `BankAccount`.

### Requirements
- constructor parameters:
  - `owner`
  - `balance`
- method `deposit(amount)`
  - add to the balance
- method `withdraw(amount)`
  - if enough money exists, subtract and return `True`
  - otherwise do not change the balance and return `False`
- method `display()`
  - return a string like:
    - `"Ari has $120."`

Create one account and test all methods.

### Reminder
This is classic object-oriented design: the object stores state and methods manage that state.

````python
# Write your answer here
class BankAccount:

  def __init__(self,owner, balance)->None:
    self.owner=owner
    self.balance=balance

  def deposit(self,amount):
    self.balance+=amount

  def withdraw(self,amount):
    if self.balance-amount >=0:
      self.balance-=amount
      return True
    else:
      return False

  def display(self):
    return f"{self.owner} has ${self.balance:,.2f}"

bob=BankAccount("Bob", 200)

bob.deposit(20)
print(bob.withdraw(100))
print(bob.display())
print(bob.withdraw(200))
print(bob.display())

````

## Task 10 — A `Classroom` and `Student`

Create two classes: `Student` and `Classroom`.

### `Student`
- constructor parameters:
  - `name`
  - `grade`

### `Classroom`
- constructor should create an empty list named `students`
- method `add_student(student_obj)`
  - add the object to the list
- method `highest_grade()`
  - return the highest grade among the students
  - if there are no students, return `None`

### Test
- create at least 3 students
- add them to a classroom
- print the highest grade

### Reminder
This is good practice with objects containing collections of other objects.

````python
# Write your answer here

class Student:

  def __init__(self, name, grade) -> None:
    self.name=name
    self.grade=grade

class Classroom:
  def __init__(self):
    self.students = []

  def add_student(self, student_obj):
    self.students.append(student_obj)

  def highest_grade(self):
    if len(self.students) ==0:
      return None
    highest=self.students[0].grade
    for student in self.students:
      if student.grade >=highest:
        highest=student.grade

    return highest

jane=Student("Jane", 90)
john=Student("John", 85)
joe=Student("Joe", 100)

science=Classroom()
science.add_student(jane)
science.add_student(john)
science.add_student(joe)
print(science.highest_grade())



````

## Task 11 — Design + object interaction challenge

Create two classes: `Player` and `InventoryManager`.

### `Player`
- constructor parameters:
  - `name`
  - `items`  (this will be a list of strings)

### `InventoryManager`
Add two methods:

1. `add_item(player_obj, item)`
   - append the item to the player's `items` list

2. `trade_first_item(player_a, player_b)`
   - if **both** players have at least one item:
     - swap their first items
     - return `True`
   - otherwise return `False`

### Test
Create two players with different item lists.
Demonstrate both methods.

### Why this is a thinking question
You must manage state stored inside objects, including a mutable list attribute.
Be careful to mutate the correct object data.

````python
# Write your answer here
class Player:

  def __init__(self, name, items) -> None:
    self.name=name
    self.items=items


class InventoryManager:

  def add_item(self,player_obj, item):
    player_obj.items.append(item)

  def trade_first_item(self, player_a, player_b):
    if len(player_a.items)>=1 and len(player_b.items)>=1:
      temp=player_b.items[0]
      player_b.items[0]=player_a.items[0]
      player_a.items[0]=temp
      return True
    return False

game1=Player("Pokemon", ["Pikachu"])
game2=Player("Football", ["Ravens", "Chiefs"])
inventory=InventoryManager()
print(game1.items)
print(game2.items)
inventory.add_item(game2, "Cowboys")
print(game1.items)
print(game2.items)
print(inventory.trade_first_item(game1, game2))
print(game1.items)
print(game2.items)
````

## Task 12 — LeetCode-style object design challenge

Create a class named `ScoreTracker`.

### Requirements
- constructor creates an empty dictionary to store player scores
- method `add_score(player_name, points)`
  - if the player is new, create the entry
  - otherwise add to the existing score
- method `top_player()`
  - return the name of the player with the highest score
  - if there are no players, return `None`
- method `top_score()`
  - return the highest score
  - if there are no players, return `None`

### Test
Add several scores for several players and show that repeated calls accumulate correctly.

### Reflection
In a comment, explain why this is a good example of a class:
- what state is being stored?
- what behaviors are being grouped with that state?

````python

# Write your answer here
class ScoreTracker:

  def __init__(self) -> None:
    self.scores={}

  def add_score(self, player_name, points):
    if player_name in self.scores:
      self.scores[player_name] += points
    else:
      self.scores[player_name]=points
  def top_player(self):
    highest_score=0
    highest_player=""
    if len(self.scores)==0:
      return None
    for name, score in self.scores.items():
      if score >= highest_score:
        highest_score=score
        highest_player=name
    return highest_player
  def top_score(self):
    highest_score=0
    if len(self.scores)==0:
      return None
    for name, score in self.scores.items():
      if score >= highest_score:
        highest_score=score
    return highest_score

tracker=ScoreTracker()
tracker.add_score("Alison", 30)
tracker.add_score("Sara", 50)
tracker.add_score("Chana", 40)
tracker.add_score("Chana", 25)
tracker.add_score("Sara", 10)

print(tracker.top_player())
print(tracker.top_score())

# This is a good example of the class as the data and methods are in one place allowing the user to read and modify the data
# The state: self.scores which is storong all the players names and totals
# The Behaviors: add_score, top_player, and top_score are all grouped together to read or edit self.scores
````

What is the difference between changing an object's contents and reassigning a local parameter variable?

When you change an object's contents, you are modifying the actual object in memory, so the change is visible outside the method. However, when you reassign a local parameter variable, you are only changing where that variable points to inside the method and it will not remain true outside of the method or wherever the variable is.
